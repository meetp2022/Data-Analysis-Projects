```python
def clean_food_waste_data(file_path):
    """
    Clean the food waste dataset from Eurostat
    """
    print("üçΩÔ∏è Cleaning Food Waste Dataset...")
    # Load data
    df = pd.read_csv(file_path)

    essential_cols = ['geo', 'TIME_PERIOD', 'nace_r2', 'OBS_VALUE']
    df_clean = df[essential_cols].copy()

    # Rename columns for clarity
    df_clean.columns = ['country', 'year', 'sector_raw', 'waste_kg_per_capita']

    sector_mapping = {
            'Primary production of food - agriculture, fishing and aquaculture': 'Primary Production',
            'Manufacturing - food industries': 'Food Manufacturing',
            'Manufacturing - beverage industries': 'Beverage Manufacturing',
            'Retail and other distribution of food': 'Retail & Distribution',
            'Restaurants and food services': 'Food Services',
            'Households': 'Households'
    }

    df_clean['sector'] = df_clean['sector_raw'].map(sector_mapping)
    df_clean = df_clean.drop('sector_raw', axis=1)

    # Handle missing values
    print(f"   Missing values before cleaning: {df_clean['waste_kg_per_capita'].isna().sum()}")
    df_clean = df_clean.dropna(subset=['waste_kg_per_capita'])

    # Convert year to integer
    df_clean['year'] = df_clean['year'].astype(int)

     # Standardize country names (EU focus)
    country_mapping = {
            'Austria': 'Austria', 'Belgium': 'Belgium', 'Bulgaria': 'Bulgaria',
            'Croatia': 'Croatia', 'Cyprus': 'Cyprus', 'Czechia': 'Czech Republic',
            'Denmark': 'Denmark', 'Estonia': 'Estonia', 'Finland': 'Finland',
            'France': 'France', 'Germany': 'Germany', 'Greece': 'Greece',
            'Hungary': 'Hungary', 'Ireland': 'Ireland', 'Italy': 'Italy',
            'Latvia': 'Latvia', 'Lithuania': 'Lithuania', 'Luxembourg': 'Luxembourg',
            'Malta': 'Malta', 'Netherlands': 'Netherlands', 'Poland': 'Poland',
            'Portugal': 'Portugal', 'Romania': 'Romania', 'Slovakia': 'Slovakia',
            'Slovenia': 'Slovenia', 'Spain': 'Spain', 'Sweden': 'Sweden'
    }

    # Filter for EU countries only
    df_clean = df_clean[df_clean['country'].isin(country_mapping.keys())].copy()
    df_clean['country'] = df_clean['country'].map(country_mapping)

    print(f"   Final food waste records: {len(df_clean)}")
    print(f"   Countries: {df_clean['country'].nunique()}")
    print(f"   Years: {sorted(df_clean['year'].unique())}")
    print(f"   Sectors: {df_clean['sector'].unique()}")

    return df_clean
```

##Food Prices

```python
def clean_food_prices_data(file_path):
    """
    Clean the food prices dataset (HICP data)
    """
    print("\nüí∞ Cleaning Food Prices Dataset...")

    # Load data
    df = pd.read_csv("/content/sample_data/Research_project/ei_cphi_m$defaultview_linear (1).csv")

    # Keep essential columns
    essential_cols = ['geo', 'TIME_PERIOD', 'OBS_VALUE']
    df_clean = df[essential_cols].copy()

    # Rename columns
    df_clean.columns = ['country', 'time_period', 'hicp_index']

    # Extract year from monthly data (e.g., "2024-10" -> 2024)
    df_clean['year'] = df_clean['time_period'].str[:4].astype(int)

    # Calculate annual average HICP for each country-year
    df_annual = df_clean.groupby(['country', 'year'])['hicp_index'].mean().reset_index()
    df_annual.columns = ['country', 'year', 'food_price_index']

    # Handle missing values
    print(f"   Missing values before cleaning: {df_annual['food_price_index'].isna().sum()}")
    df_annual = df_annual.dropna(subset=['food_price_index'])

    # Standardize country names to match food waste dataset
    country_mapping = {
        'Austria': 'Austria', 'Belgium': 'Belgium', 'Bulgaria': 'Bulgaria',
        'Croatia': 'Croatia', 'Cyprus': 'Cyprus', 'Czechia': 'Czech Republic',
        'Denmark': 'Denmark', 'Estonia': 'Estonia', 'Finland': 'Finland',
        'France': 'France', 'Germany': 'Germany', 'Greece': 'Greece',
        'Hungary': 'Hungary', 'Ireland': 'Ireland', 'Italy': 'Italy',
        'Latvia': 'Latvia', 'Lithuania': 'Lithuania', 'Luxembourg': 'Luxembourg',
        'Malta': 'Malta', 'Netherlands': 'Netherlands', 'Poland': 'Poland',
        'Portugal': 'Portugal', 'Romania': 'Romania', 'Slovakia': 'Slovakia',
        'Slovenia': 'Slovenia', 'Spain': 'Spain', 'Sweden': 'Sweden'
    }

    # Filter for EU countries
    df_annual = df_annual[df_annual['country'].isin(country_mapping.keys())].copy()
    df_annual['country'] = df_annual['country'].map(country_mapping)

    print(f"   Final food price records: {len(df_annual)}")
    print(f"   Countries: {df_annual['country'].nunique()}")
    print(f"   Years: {sorted(df_annual['year'].unique())}")

    return df_annual
```

##Merge Dataframes

```python
def merge_all_datasets(food_waste_df, food_prices_df, gdp_df, population_df):
    """
    Merge all cleaned datasets and handle final data issues
    """
    print("\nüîó Merging All Datasets...")

    # Start with food waste as base (most specific to your project)
    merged_df = food_waste_df.copy()

    # Merge food prices
    merged_df = merged_df.merge(
        food_prices_df,
        on=['country', 'year'],
        how='left',
        indicator='_merge_prices'
    )

    # Merge GDP
    merged_df = merged_df.merge(
        gdp_df,
        on=['country', 'year'],
        how='left',
        indicator='_merge_gdp'
    )

    # Merge population
    merged_df = merged_df.merge(
        population_df,
        on=['country', 'year'],
        how='left',
        indicator='_merge_population'
    )

    # Check merge quality
    print("\nüìä Merge Quality Assessment:")
    print(f"   Total records after merge: {len(merged_df)}")
    print(f"   Records with food prices: {(merged_df['food_price_index'].notna()).sum()}")
    print(f"   Records with GDP data: {(merged_df['gdp_index'].notna()).sum()}")
    print(f"   Records with population: {(merged_df['population'].notna()).sum()}")

    # Handle missing values in context indicators
    # Option 1: Drop rows with missing context data (recommended for small dataset)
    # Option 2: Simple imputation (use if too much data loss)

    # Check missing data patterns
    missing_analysis = merged_df.isnull().sum()
    print(f"\nüîç Missing Data Analysis:")
    for col in ['food_price_index', 'gdp_index', 'population']:
        missing_count = missing_analysis[col]
        missing_pct = (missing_count / len(merged_df)) * 100
        print(f"   {col}: {missing_count} missing ({missing_pct:.1f}%)")

    # Strategy: Drop rows with missing context indicators (since we need them for correlation)
    df_final = merged_df.dropna(subset=['food_price_index', 'gdp_index', 'population']).copy()

    # Remove merge indicator columns
    df_final = df_final.drop([col for col in df_final.columns if '_merge' in col], axis=1)

    print(f"\n‚úÖ Final clean dataset: {len(df_final)} records")
    print(f"   Countries: {df_final['country'].nunique()}")
    print(f"   Years: {sorted(df_final['year'].unique())}")
    print(f"   Sectors: {df_final['sector'].nunique()}")

    return df_final
```

##Data Summary

```python
def generate_data_summary(df):
    """
    Generate summary statistics for the clean dataset
    """
    print("\nüìà Data Summary Statistics:")
    print("="*50)

    # Basic stats
    print(f"Dataset shape: {df.shape}")
    print(f"Time period: {df['year'].min()} - {df['year'].max()}")
    print(f"Countries: {df['country'].nunique()}")
    print(f"Sectors: {df['sector'].nunique()}")

    print(f"\nCountry coverage:")
    country_counts = df['country'].value_counts()
    print(f"   Most data: {country_counts.index[0]} ({country_counts.iloc[0]} records)")
    print(f"   Least data: {country_counts.index[-1]} ({country_counts.iloc[-1]} records)")

    print(f"\nFood waste statistics (kg per capita):")
    print(f"   Mean: {df['waste_kg_per_capita'].mean():.1f}")
    print(f"   Median: {df['waste_kg_per_capita'].median():.1f}")
    print(f"   Range: {df['waste_kg_per_capita'].min():.1f} - {df['waste_kg_per_capita'].max():.1f}")

    print(f"\nTop 5 sectors by average waste:")
    sector_avg = df.groupby('sector')['waste_kg_per_capita'].mean().sort_values(ascending=False)
    for i, (sector, avg_waste) in enumerate(sector_avg.head().items()):
        print(f"   {i+1}. {sector}: {avg_waste:.1f} kg/capita")

    return df.describe()
```

```python
# Main execution function
def main():
    """
    Execute the complete data cleaning pipeline
    """
    print("üöÄ Starting EU Food Waste Data Cleaning Pipeline")
    print("="*60)

    # File paths (adjust these to your actual file names)
    file_paths = {
        'food_waste': '/content/sample_data/Research_project/env_wasfw$defaultview_linear (1).csv',
        'food_prices': '/content/sample_data/Research_project/ei_cphi_m$defaultview_linear (1).csv',
        'gdp': '/content/sample_data/Research_project/naidq_10_gdp$defaultview_linear (1).csv',
        'population': '/content/sample_data/Research_project/tps00001_page_linear (1).csv'
    }

    try:
        # Clean individual datasets
        food_waste_clean = clean_food_waste_data(file_paths['food_waste'])
        food_prices_clean = clean_food_prices_data(file_paths['food_prices'])
        gdp_clean = clean_gdp_data(file_paths['gdp'])
        population_clean = clean_population_data(file_paths['population'])

        # Merge all datasets
        final_dataset = merge_all_datasets(
            food_waste_clean,
            food_prices_clean,
            gdp_clean,
            population_clean
        )

        # Generate summary
        summary_stats = generate_data_summary(final_dataset)

        # Save cleaned data
        output_filename = 'eu_food_waste_cleaned.csv'
        final_dataset.to_csv(output_filename, index=False)
        print(f"\nüíæ Cleaned dataset saved as: {output_filename}")

        # Save individual cleaned datasets for reference
        food_waste_clean.to_csv('food_waste_cleaned.csv', index=False)
        food_prices_clean.to_csv('food_prices_cleaned.csv', index=False)
        gdp_clean.to_csv('gdp_cleaned.csv', index=False)
        population_clean.to_csv('population_cleaned.csv', index=False)


        print("\nüéØ Data cleaning completed successfully!")
        print("Ready for Step 3 - Analysis!")

        return final_dataset, summary_stats

    except FileNotFoundError as e:
        print(f"‚ùå Error: Could not find file {e}")
        print("Please check your file names and paths")
        return None, None # Return None if file not found
    except Exception as e:
        print(f"‚ùå Unexpected error: {e}")
        return None, None # Return None for any other exception
```

```python
# Additional utility functions
def check_data_quality(df):
    """
    Check data quality issues
    """
    print("\nüîç Data Quality Check:")

    # Check for duplicates
    duplicates = df.duplicated(subset=['country', 'year', 'sector']).sum()
    print(f"   Duplicate records: {duplicates}")

    # Check for outliers in food waste
    Q1 = df['waste_kg_per_capita'].quantile(0.25)
    Q3 = df['waste_kg_per_capita'].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df['waste_kg_per_capita'] < (Q1 - 1.5 * IQR)) |
                (df['waste_kg_per_capita'] > (Q3 + 1.5 * IQR))).sum()
    print(f"   Potential outliers in food waste: {outliers}")

    # Check year coverage by country
    country_year_coverage = df.groupby('country')['year'].nunique()
    incomplete_countries = country_year_coverage[country_year_coverage < 3].index.tolist()
    if incomplete_countries:
        print(f"   Countries with incomplete year coverage: {incomplete_countries}")

    return df
```
